```python
# import the necessary packages
from loader_util.utils.simple_obj_det import image_pyramid, sliding_window, classify_batch
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.applications import imagenet_utils
from imutils.object_detection import non_max_suppression
import numpy as np
import time
import cv2
```

```python
image_path = r"C:\Users\mhasa\PycharmProjects\DL4CV\keras_dir\biblio2\p13_object_detection\collection_cars.jpg"
image_path = r"C:\Users\mhasa\PycharmProjects\DL4CV\keras_dir\biblio2\p13_object_detection\cars_on_road.jpg"
args = {
    "image_path": image_path,
    "confidence": 0.5
}
```

```python
# init the variables
input_size = (350, 350)
pyramid_scale = 1.50
window_step = 16
roi_size = (120, 120)
batch_size = 64
```

```python
# load the network weights from disk
print(f"[INFO] loading network......")
model = ResNet50(weights="imagenet", include_top=True)
```

```python
# init the object detection dictionary which maps class labels to their predicted bounding box
labels = {}

# load the input image and grab dimensions
original_image = cv2.imread(args["image_path"])
h, w = original_image.shape[:2]

# resize the image
resized_image = cv2.resize(original_image, input_size, interpolation=cv2.INTER_CUBIC)

# init the batch ROIs and x,y coordinates
batch_rois = None
batch_locs = []
```

```python
print(f"[INFO] starting to detect objects......")
start = time.time()

# loop over the image pyramid layers
for image in image_pyramid(resized_image, scale=pyramid_scale, minsize=roi_size):
    # loop over the rois for a certain pyramid layer
    for x, y, roi in sliding_window(image, step=window_step, ws=roi_size):

        # take the roi and preprocess it since we will input to imagenet trained model
        roi = img_to_array(roi)
        roi = np.expand_dims(roi, axis=0)
        roi = imagenet_utils.preprocess_input(roi)

        if batch_rois is None:
            batch_rois = roi
        else:
            batch_rois = np.vstack([batch_rois, roi])

        batch_locs.append((x, y))

        # if batch is fullsize then classify the rois
        if len(batch_rois) == batch_size:
            labels = classify_batch(model, batch_rois=batch_rois, batch_locs=batch_locs, labels=labels, min_prob=args["confidence"])

            batch_rois = None
            batch_locs = []

# check to see any rois are reamining or not
if batch_rois is not None:
    labels = classify_batch(model, batch_rois, batch_locs, labels, args["confidence"])

# show how long detection took
end = time.time()
print(f"[INFO] detections took: {end - start: 0.4f} seconds......")
```

```python
# loop over the labels for each detected label in the image
for k in labels.keys():
    cloned_image = resized_image.copy() # type: np.ndarray

    # loop over all the bounding boxes for the label and draw them in the image
    for box, prob in labels[k]:
        x1, y1, x2, y2 = box
        cv2.rectangle(cloned_image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # now show the image
    import matplotlib.pyplot as plt
    plt.imshow(cloned_image[:,:,::-1])
```
