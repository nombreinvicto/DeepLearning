```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.axes import Axes

sns.set()
```

```python
# trusting on numpy broadcasting !!!
vect_2d = np.array([1, 2]).reshape(2, -1)
scalar = np.random.randn(10)
```

```python
vect_2d[0][0]
```

```python
all_vectors = vect_2d * scalar
```

```python
# now plot all the vectors
sns.set(font_scale=1)
f, ax = plt.subplots(1, 1, figsize=(12, 8))  # type: Figure, Axes
sns.lineplot(x=[0, vect_2d[0][0]], y=[0, vect_2d[1][0]], ax=ax, color="black", linewidth=2)
for i in range(all_vectors.shape[1]):
    vector = all_vectors[:, i]
    sns.lineplot(x=[0, vector[0]], y=[0, vector[1]], ax=ax, color="red", linewidth=1)
    ax.arrow(x=0, y=0, dx=vector[0], dy=vector[1], width=0.09)
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.set_xlabel("X")
ax.set_ylabel("Y");
```
