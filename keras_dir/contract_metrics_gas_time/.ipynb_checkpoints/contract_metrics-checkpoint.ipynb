```python
def run_bootstrap(data, samples=1000, size=10):
    time_list = []
    for i in range(samples):
        s = np.random.choice(data, size=size, replace=True)
        time_list.append(np.mean(s))
    return time_list    
```

```python
rinkeby_bs_po_time = run_bootstrap(rinkeby.purchaseOrderTime)
rinkeby_bs_mo_time = run_bootstrap(rinkeby.makeOrderTime)
rinkeby_bs_so_time = run_bootstrap(rinkeby.sourceTime)
rinkeby_bs_ge_time = run_bootstrap(rinkeby.genTime)
```

```python
f, ax = plt.subplots(2,2, figsize=(15,12))
sns.distplot(rinkeby_bs_po_time, ax=ax[0][0])
ax[0][0].set_title("initiatePurchaseOrder")
ax[0][0].set_xlabel("Time in sec")
```
