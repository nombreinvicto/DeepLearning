```python
from gensim.models import KeyedVectors
```

```python
dbBase = r"/home/mhasan3/Desktop/WorkFolder/NLP_Data_Repo/"
outBase = r"/home/mhasan3/Desktop/WorkFolder/NLP_Output/"
filename = f"{dbBase}/GoogleNews-vectors-negative300.bin"
word2vec_model = KeyedVectors.load_word2vec_format(fname=filename, binary=True)
```

```python
helper1 = "spring"
helper2 = "helix"

main_word = "coil"
topn = 1000
result = word2vec_model.most_similar(positive=[helper1, helper2, main_word],topn=topn)
```

```python
result
```

```python
output_file = f"{outBase}//{main_word}.txt"
with open(file=output_file, mode='a') as file:
    for line in result:
        line = line[0]
        line = line.replace("_", " ")
        file.write(line)
        file.write("\n")
```
